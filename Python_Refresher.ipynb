```python
# Program to calculate the area and perimeter of a square

# Function to calculate area of a square
def find_area(side):
    return side * side

# Function to calculate perimeter of a square
def find_perimeter(side):
    return 4 * side

# Main part of the program
def main():
    # Ask the user to enter the side length of the square
    side_length = float(input("Enter the side length of the square: "))

    # Call functions to calculate area and perimeter
    area = find_area(side_length)
    perimeter = find_perimeter(side_length)

    # Display the results
    print("Area of the square:", area)
    print("Perimeter of the square:", perimeter)

# Run the main function
main()
```

```python
# Program to check if a character is uppercase or lowercase

# Function to check case
def check_case(char):
    if char.isupper():
        return "The character is UPPERCASE."
    elif char.islower():
        return "The character is lowercase."
    else:
        return "The input is not a letter."

# Main function
def main():
    # Ask the user to enter a character
    char = input("Enter a single character: ")

    # Make sure the user enters only one character
    if len(char) != 1:
        print("Please enter only ONE character.")
    else:
        # Call the function and display result
        result = check_case(char)
        print(result)

# Run the program
main()
```

```python
# Program to convert pseudocode to python code

def main():
    # Initialize values
    x = 0
    y = 20

    # Repeat until y is less than 6
    while y >= 6:
        y = y - 4           # subtract 4 from y
        x = x + (2 / y)     # add 2/y to x

    # Display result
    print("Final value of x is:", x)

# Run the program
main()
```

```python

# Program to input 5 values, store them in an array (list) and calculate the average

def main():
    numbers = []  # empty list to store user inputs

    print("Please enter 5 numbers:")

    # Loop 5 times
    for i in range(5):
        value = float(input("Enter number " + str(i+1) + ": "))
        numbers.append(value)  # add the number to the list

    # Calculate average
    total = sum(numbers)
    average = total / len(numbers)

    # Display results
    print("The numbers you entered are:", numbers)
    print("The average is:", average)

# Run the program
main()
```
